```python
import torch
import torch.nn as nn
from collections import OrderedDict
import matplotlib.pyplot as plt
```

```python
model1 = nn.Sequential(OrderedDict([
   ('hidden', nn.Linear(2, 8)),
   ('relu', nn.ReLU()),
   ('output', nn.Linear(8, 2)),
   ('sigmoid', nn.Sigmoid())
]))
model2 = nn.Sequential(OrderedDict([
   ('hidden', nn.Linear(2, 8)),
   ('tanh', nn.Tanh()),
   ('output', nn.Linear(8, 2)),
   ('sigmoid', nn.Sigmoid())
]))
model3 = nn.Sequential(OrderedDict([
   ('hidden', nn.Linear(2, 4)),
   ('tanh', nn.Tanh()),
   ('hidden2', nn.Linear(4, 8)),
   ('relu', nn.ReLU()),
   ('output', nn.Linear(8, 2)),
   ('sigmoid', nn.Sigmoid())
]))
models = [model1, model2, model3]
```

```python
for model in models:
    print(model)
```

```python
data_in = torch.tensor([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=torch.float32)
print(data_in)
```

```python
data_target = torch.tensor([[0, 0], [0, 1], [0, 1], [1, 0]], dtype=torch.float32)
print(data_target)
```

```python
criterion = nn.BCELoss()
optimizers = [ torch.optim.SGD(model.parameters(), lr=0.1) for model in models ]
```

```python
# your code here
# Train the model

general_results = []
all_models_losses = [] 

for i, model in enumerate(models):
    num_epochs = 1000
    losses = [] 
    optimizer = optimizers[i]
    
    for epoch in range(num_epochs):
        output = model(data_in)
        loss = criterion(output, data_target)
    
        optimizer.zero_grad()
        loss.backward()  
        optimizer.step() 
        
        losses.append(loss.item())

    accuracy = ((output > 0.5).float() == data_target).float().mean().item()

    print(f"Model {i+1}: Loss = {loss.item()}, Accuracy = {accuracy}")
    
    general_results.append((i, loss.item(), accuracy))
    all_models_losses.append(losses) 
```

```python
# your code here
# visualize the resuts

plt.figure(figsize=(10, 6))

for i, losses in enumerate(all_models_losses):
    plt.plot(losses, label=f'Model {i+1}')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Function Convergence')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# your code here
# print model wights

best_model_index = max(general_results, key=lambda x: x[2])[0]
print(f"Best Model: Model {best_model_index + 1}")
print(f"Best Model Weights: {models[best_model_index].state_dict()}")
```
